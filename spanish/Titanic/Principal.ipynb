<h1>Análisis de la tripulación del titanic</h1>

El Titanic fue un transatlántico británico, el mayor barco de pasajeros del mundo al finalizar su construcción. Este se hundió el 15 de abril del 1912 durante su viaje inaugural desde Southampton a Nueva York. En este trágico suceso se dice que murieon 1514 personas de 2223 que iban a bordo, lo que lo convierte en uno de los mayores naufragios de la historia.

![alt text](https://upload.wikimedia.org/wikipedia/commons/f/fd/RMS_Titanic_3.jpg)

Entre sus pasajeros estaban algunas de las personas más ricas del mundo, además de cientos de inmigrantes irlandeses, británicos y escandinavos que iban en busca de una mejor vida en Norteamérica.

El Titanic recaló en Cherburgo (Francia) y en Queenstown (Irlanda) antes de poner proa al océano Atlántico.

El Titanic chocó contra un iceberg. La colisión abrió varias planchas del casco en su lado de estribor bajo la línea de flotación. Durante dos horas y media el barco se fue hundiendo gradualmente por su parte delantera mientras la popa se elevaba y en este tiempo varios cientos de pasajeros y tripulantes fueron evacuados en los botes salvavidas. Solo se portaban botes salvavidas para 1178 pasajeros, poco más de la mitad de los que iban a bordo en su viaje inaugural y un tercio de su capacidad total.

Ahora vamos a buscar correlaciones 

```python
def plot_correlation_map( df ):
    corr = data.corr()
    _ , ax = plt.subplots( figsize =( 12 , 10 ) )
    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )
    _ = sns.heatmap(
        corr, 
        cmap = cmap,
        square=True, 
        cbar_kws={ 'shrink' : .9 }, 
        ax=ax, 
        annot = True, 
        annot_kws = { 'fontsize' : 12 }
    )
```

```python
plot_correlation_map(data)
```

Vamos a intentar demostrar la hipótesis de que la gente de mayor clase social (clase A) sobrevivía con mayor frecuencia.

```python
sns.countplot(data['Pclass'], hue=data['Survived'])
```

Vemos como la mayor parte de la gente de clase A sobrevivía. La gente de clase B mitad y mitad y por la último la clase C no solía sobrevivir. Una posible respuesta a esto es porque las cabinas de los pasajeros de clase C se encontraban en la parte inferior del barco, con lo cuál al chocar el iceber en la parte inferior provocó que mucha gente se quedará atrapada bajo el agua. Otra posible respuesta, es que al ser gente de menos recursos no podía optar en muchos casos al bote salvavidas.

```python

for i in range(3):
    passengerClass = data[data.Pclass == (i+1)].Survived.value_counts()
    print ("Class of Passenger: ", (i+1), " - percent of survived: ", (passengerClass[1] / passengerClass.sum()) * 100)
```

Ahora vamos a intentar demostrar la hipótesis de que la gente de mayor clase social es de una edad avanzada:

```python
dataSortedByAge = data.sort_values(by='Age')[data.Age.notnull()]
dataAgeFare = dataSortedByAge.set_index(dataSortedByAge['Age'])
dataAgeFare = dataAgeFare[['Fare']]
dataAgeFare.plot()
```

Vemos que mi hipótesis no es cierta, ya que la mayoria de la gente joven pagó más dinero que la gente mayor a excepción de dos picos importantes a entre los 35-40 años.

Ahora vamos a mirar el número de supervivientes según el sexo de los pasajeros:

```python
sns.countplot(data['Sex'], hue=data['Survived'])
```

Vemos que murieron que sobrevivieron más mujeres que hombres, esto puede ser debido a la estricta regla de "mujeres y niños primero". Ahora miraremos dicha proporción para gente de primera clase exclusivamente:

```python
sns.countplot(data[data.Pclass == 1]['Sex'], hue=data['Survived'])
```

Vemos que la mayor parte de las mujeres sobrevivieron y de los hombres una parte importante también sobrevivió.

Ahora comprobaremos la probabilidad de sobrevivir si: estamos solteros, pequeña familia y grande familia:

```python
data['Family']= data['Parch']+ data['SibSp']+1
data.loc[data["Family"] == 1, "FamilySize"] = 'singleton'
data.loc[(data["Family"] > 1)  &  (data["Family"] < 5) , "FamilySize"] = 'small'
data.loc[data["Family"] >4, "FamilySize"] = 'large'
sns.countplot(data['FamilySize'],hue=data['Survived'])
```

Vemos que la mayoría de gente que iba sola al Titanic, falleció, en cambio los pasajeros con una pequeña familia se aseguraba que algún familiar pudiera sobrevivir.

Ahora vamos a comprobar la media y desviación standard de la tarifa para supervivientes y para no supervivientes:

```python


fare_not_survived = data["Fare"][data["Survived"] == 0]
fare_survived     = data["Fare"][data["Survived"] == 1]

average_fare = pd.DataFrame([fare_not_survived.mean(), fare_survived.mean()])
std_fare      = pd.DataFrame([fare_not_survived.std(), fare_survived.std()])
average_fare
```
